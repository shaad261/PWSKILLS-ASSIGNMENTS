```python
#Q1 Matplotlib is a cross-platform, data visualization and graphical plotting library (histograms, scatter plots, bar charts, etc) for Python 
#A Python matplotlib script is structured so that a few lines of code are all that is required in most instances to generate a visual data plot.
```

```python
#Q2 A scatter plot uses dots to represent values for two different numeric variables.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.figure(figsize=(6,2))
plt.scatter(x,y , c = '#223BDC',alpha=.9,marker='X')
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my x vs y")
```

```python
#Q3 The subplot() function takes three arguments that describes the layout of the figure.
# The layout is organized in rows and columns, which are represented by the first and second argument.
#The third argument represents the index of the current plot.
#plot 1:
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(4, 1, 1)
plt.plot(x,y)
#plot 2:
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(4, 1, 2)
plt.plot(x,y)
#plot 3:
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(4, 1, 3)
plt.plot(x,y)
#plot 4:
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(4, 1, 4)
plt.plot(x,y)
```

```python
#Q4 A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. 
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize=(5,3))
plt.bar(company,profit,color = '#7E10D5')
plt.xlabel("representing company")
plt.ylabel("representing profit")
plt.title("this is my bar plot")
```

```python
#Q5 A Box Plot is the visual representation of the statistical five number summary of a given data set.
# A Five Number Summary includes:
# Minimum
# First Quartile
# Median (Second Quartile)
# Third Quartile
# Maximum
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(data)
plt.show()
```
